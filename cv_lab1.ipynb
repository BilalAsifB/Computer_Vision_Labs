```python
class GroceryList:
    def __init__(self):
        self.items = {}

    def add_item(self, item):
        if item in self.items:
            self.items[item] += 1
        else:
            self.items[item] = 1

    def remove_item(self, item):
        if item in self.items:
            if self.items[item] > 1:
                self.items[item] -= 1
            else:
                del self.items[item]

    def display_list(self):
        if not self.items:
            print("Grocery list is empty!")
            return

        table = tabulate(self.items.items(), headers=["Item", "Quantity"], tablefmt="fancy_grid")
        print(f"{' '*6}Grocery List{' '*6}")
        print(table)
```

```python
g = GroceryList()
g.add_item("Milk")
g.add_item("Eggs")
g.add_item("Milk")
g.display_list()
```

```python
class StudentRecordSystem:
    def __init__(self):
        self.records = {}

    def add_student(self, student_id, name, grade):
        if student_id in self.records:
            print(f"Student ID {student_id} already exists.")
        else:
            self.records[student_id] = {"Name": name, "Grade": grade}
            print(f"Added student {name} (ID: {student_id})")

    def update_grade(self, student_id, new_grade):
        if student_id in self.records:
            self.records[student_id]["Grade"] = new_grade
            print(f"Updated grade for {self.records[student_id]['Name']} (ID: {student_id})")
        else:
            print("Student not found.")

    def remove_student(self, student_id):
        if student_id in self.records:
            removed = self.records.pop(student_id)
            print(f"Removed student {removed['Name']} (ID: {student_id})")
        else:
            print("Student not found.")

    def display_records(self):
        if not self.records:
            print("No student records available.")
            return

        table = []
        for sid, info in self.records.items():
            table.append([sid, info["Name"], info["Grade"]])

        print(f"\n{' '*11}Student Records{' '*11}")
        print(tabulate(table, headers=["Student ID", "Name", "Grade"], tablefmt="fancy_grid"), '\n')
```

```python
srs = StudentRecordSystem()
srs.add_student("101", "Alice", "A")
srs.add_student("102", "Bob", "B+")
srs.add_student("103", "Charlie", "A-")

srs.display_records()

srs.update_grade("102", "A")
srs.remove_student("103")

srs.display_records()
```

```python
img = cv2.imread("/content/sample_data/image.jpeg", cv2.IMREAD_GRAYSCALE)
_, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

(h, w) = thresh.shape
M = cv2.getRotationMatrix2D((w//2, h//2), 60, 1)
rotated = cv2.warpAffine(thresh, M, (w,h))

plt.figure(figsize=(15,5))
plt.subplot(1,2,1); plt.imshow(thresh, cmap="gray"); plt.title("Threshold")
plt.subplot(1,2,2); plt.imshow(rotated, cmap="gray"); plt.title("Rotated 60°")
plt.show()
```

```python
img1 = cv2.imread("/content/sample_data/image.jpeg")
img2 = cv2.imread("/content/sample_data/image2.jpeg")
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

blend = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)

gray = cv2.cvtColor(blend, cv2.COLOR_BGR2GRAY)
equalized = cv2.equalizeHist(gray)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(blend, cv2.COLOR_BGR2RGB)); plt.title("Blended")
plt.subplot(1,2,2); plt.imshow(equalized, cmap="gray"); plt.title("Histogram Equalized")
plt.show()
```

```python
img1 = np.zeros((300,300), dtype="uint8")
img2 = np.zeros((300,300), dtype="uint8")
cv2.rectangle(img1, (50,50), (250,250), 255, -1)
cv2.circle(img2, (150,150), 100, 255, -1)

bitwise_and = cv2.bitwise_and(img1, img2)
bitwise_or = cv2.bitwise_or(img1, img2)
bitwise_xor = cv2.bitwise_xor(img1, img2)
bitwise_not = cv2.bitwise_not(img1)

plt.figure(figsize=(10,10))
plt.subplot(2,2,1); plt.imshow(bitwise_and, cmap="gray"); plt.title("AND")
plt.subplot(2,2,2); plt.imshow(bitwise_or, cmap="gray"); plt.title("OR")
plt.subplot(2,2,3); plt.imshow(bitwise_xor, cmap="gray"); plt.title("XOR")
plt.subplot(2,2,4); plt.imshow(bitwise_not, cmap="gray"); plt.title("NOT")
plt.show()
```

```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate

# Load grayscale image
img = cv2.imread("/content/sample_data/image2.jpeg", cv2.IMREAD_GRAYSCALE)
df = pd.DataFrame(img)

# Compute statistics
stats = df.describe().T

print(f"{' '*28}Image Statistics{' '*28}")
print(tabulate(stats, headers="keys", tablefmt="fancy_grid"))

mask = cv2.bitwise_and(img, img, mask=(img > 128).astype("uint8") * 255)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1); plt.imshow(img, cmap="gray"); plt.title("Original")
plt.subplot(1,2,2); plt.imshow(mask, cmap="gray"); plt.title("Masked (AND)")
plt.show()
```
